### Task 4: Most Preferred Investment Avenue

Objective: Identify the most preferred investment avenue.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\Data_set 2 - Copy.csv")
```

```python

# Columns representing investment avenues
investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures',
                      'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']

# Calculating the total preference for each avenue
investment_totals = df[investment_columns].sum()

# Finding the most preferred investment avenue
most_preferred = investment_totals.idxmax()
most_preferred_value = investment_totals.max()

print("Investment Totals:")
print(investment_totals)
print(f"\nThe most preferred investment avenue is {most_preferred} with a total score of {most_preferred_value}.")
```

```python
# Step 1: Analyzing Investment Avenues (single column)
print("Unique values in 'Avenue' column:", df['Avenue'].unique())

# Step 2: Frequency Analysis
avenue_counts = df['Avenue'].value_counts()
print("\nInvestment Avenue Frequencies:\n")
print(avenue_counts)

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(x=avenue_counts.values, y=avenue_counts.index, palette='Blues')
plt.title('Most Preferred Investment Avenues', fontsize=16)
plt.xlabel('Number of Participants', fontsize=14)
plt.ylabel('Investment Avenue', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
