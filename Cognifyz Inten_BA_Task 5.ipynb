### Objective: Analyze and summarize reasons for investment choices.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("D:\Data_set 2 - Copy.csv")

# Analyzing Purpose column
purpose_counts = df['Purpose'].value_counts()

# Results
print("Investment Purposes:")
print(purpose_counts)

# Plot 
purpose_counts.plot(kind='bar', color='skyblue', title='Investment Purposes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
